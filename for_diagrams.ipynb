```python
import networkx as nx
import numpy as np
import pylab as pl

%matplotlib inline
pl.style.use('latexplot')
```

```python
G = nx.grid_2d_graph(5, 5)
```

```python
pos=nx.spring_layout(G) # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)

nx.draw_networkx_nodes(G, pos, nodelist = [(1, 1)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(1, 2)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(1, 3)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(2, 1)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(2, 2)], node_color='r')
nx.draw_networkx_nodes(G, pos, nodelist = [(2, 3)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(3, 1)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(3, 2)], node_color='k')
nx.draw_networkx_nodes(G, pos, nodelist = [(3, 3)], node_color='k')

nx.draw_networkx_edges(G, pos, edgelist=[((0, 0),(1, 1))], edge_color = 'gray', style = '--')
```
